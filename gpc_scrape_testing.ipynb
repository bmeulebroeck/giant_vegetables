```python
from bs4 import BeautifulSoup
import requests
import os
import pandas as pd
```

```python
# URL to 2021 results for testing
url = 'http://www.bigpumpkins.com/WeighoffResultsGPC.aspx?c=W&y=2021'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
```

```python
results = soup.find_all('table', class_='ReportResults')
# print(results)
```

```python
#Isolate the table with the results
results_table = soup.find("table", attrs={"class":"ReportResults"})

#Column headings are in the thead, so need to seperate that from the actual results data for further processing
results_table_columns = results_table.thead.find_all("th")
results_table_data = results_table.tbody.find_all("tr")


#Get the list of column headings for later
columns = []
for th in results_table_columns:
    # columns.append(th.text.replace('\n', ' ').strip())
    columns.append(th.text)

# print(results_table_columns)
print(columns)
# print(results_table_data)
```
